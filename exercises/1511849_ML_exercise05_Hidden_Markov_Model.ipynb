```
import pandas as pd
import numpy as np
eps = np.finfo(float).eps
```

```
trans_prob = pd.DataFrame({'TV':   [0.4, 0.6 , 0.7 , 0.3 ],
                           'Pub':  [0.3, 0.05, 0.05, 0.4 ],
                           'Party':[0.1, 0.1 , 0.05, 0.25],
                           'Study':[0.2, 0.25, 0.2 , 0.05]},
                          index = ['TV', 'Pub', 'Party','Study'],
                          columns = ['TV', 'Pub', 'Party','Study'])

emission_prob = pd.DataFrame({'TV':   [0.2, 0.1 , 0.2, 0.5 ],
                              'Pub':  [0.4, 0.2 , 0.1, 0.3 ],
                              'Party':[0.3, 0.4 , 0.2, 0.1 ],
                              'Study':[0.3, 0.05, 0.3, 0.35]},
                             index = ['Tired', 'Hungover', 'Scared','Fine'],
                             columns = ['TV', 'Pub', 'Party','Study'])
trans_prob
```

```
emission_prob
```

```
start_prob = [0.25, 0.25, 0.25, 0.25]
obsv = ['Hungover' , 'Scared', 'Tired']
```

```
def forward(start_probability, observation):
  layer_input = start_probability
  for obsv in observation:
    x = emission_prob.loc[[obsv]].values   
    temp = layer_input * x    
    print(obsv, '\t' ,temp.flatten())
    layer_input = np.matmul(temp, trans_prob.values)
    
```

```
forward(start_prob, obsv)
```
